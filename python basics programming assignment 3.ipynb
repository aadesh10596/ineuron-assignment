#### 2. Write a Python Program to Check if a Number is Odd or Even?

```python
a = int(input("Enter any number"))

if a % 2 == 0:
    print("number {} is even number".format(a))
else:
    print("number {} is odd number".format(a))
```

#### 3. Write a Python Program to Check Leap Year?

```python
a = int(input("Enter year in YYYY format "))

if a % 100 == 0:
    if a % 400 == 0:
        print("Year {} is leap year".format(a))
    else:
        print("Year {} is not leap year".format(a))
else:
    if a % 4 == 0:
        print("Year {} is leap year".format(a))
    else:
        print("Year {} is not leap year".format(a))
```

#### 4. Write a Python Program to Check Prime Number?

```python
a = int(input("enter any number "))

if a > 1:
    for i in range(2, int(a/2)+1):
        if (a % i) == 0:
            print("{} is not a prime".format(a))
            break
    else:
        print("{} is a prime".format(a))
else:
    if a == 1:
        print("{} is neither prime nor composite".format(a))
    else:
        print("{} is a prime".format(a))
```
